```python
 # pennylane
import pennylane as qml
from pennylane import numpy as np

# dataset
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split

import matplotlib.pyplot as plt
from skimage.transform import resize
```

## AmplitudeEmbedding
https://docs.pennylane.ai/en/stable/code/api/pennylane.AmplitudeEmbedding.html


```python
dev = qml.device('default.qubit', wires=4)

@qml.qnode(dev)
def circuit(f=None):
    qml.AmplitudeEmbedding(features=f, wires=range(4))
    return qml.state()

# Normalize the features array
features = np.array([1/2] * 16)
features = features / np.linalg.norm(features)  # Normalize the array

fig = qml.draw_mpl(circuit)(f=features)
plt.show()
```

## AngleEmbedding
https://docs.pennylane.ai/en/stable/code/api/pennylane.AngleEmbedding.html

```python
dev = qml.device("default.qubit", wires=4)

# Define the QNode
@qml.qnode(dev)
def circuit(feature_vector):
    qml.AngleEmbedding(features=feature_vector, wires=range(4), rotation='Z')
    qml.Hadamard(0)
    return qml.probs(wires=range(3))

# feature vector
X = [1, 2, 3, 4]

fig = qml.draw_mpl(circuit)(X)
plt.show()
```

### Basis Embedding

https://docs.pennylane.ai/en/stable/code/api/pennylane.BasisEmbedding.html

```python
dev = qml.device('default.qubit', wires=4)

@qml.qnode(dev)
def circuit(feature_vector):
    qml.BasisEmbedding(features=feature_vector, wires=range(4))
    return qml.state()

X = [1,1,1,1]
fig = qml.draw_mpl(circuit)(X)
plt.show()
```
